GUI

```python
import tkinter as tk
from tkinter import messagebox, ttk
import pyodbc

server = "saif\\SQLEXPRESS"
Database = "e_commerce"
Driver = "{ODBC Driver 17 for SQL Server}" 

try:
    conn_str = f"DRIVER={Driver};SERVER={server};DATABASE={Database};Trusted_Connection=yes;"
    connection = pyodbc.connect(conn_str)
    cursor = connection.cursor()
    print("Database connected successfully")
except pyodbc.Error as e:
    print("Database connection failed:", e)
    exit()


def retrieve_data():
    try:
        table_name = entry_table_name.get()

        if not table_name:
            messagebox.showwarning("Input Error", "Table name is required!")
            return


        cursor.execute(f"SELECT * FROM {table_name}")
        rows = cursor.fetchall()

        if not rows:
            messagebox.showinfo("No Data", f"No data found in table '{table_name}'.")
            return


        data_window = tk.Toplevel()
        data_window.title(f"Data from {table_name}")

        text = tk.Text(data_window, wrap="word", width=80, height=20)
        text.pack()

        column_headers = [desc[0] for desc in cursor.description]
        text.insert("1.0", "\t".join(column_headers) + "\n")
        text.insert("2.0", "-" * 80 + "\n")

    
        for row in rows:
            text.insert("end", "\t".join(map(str, row)) + "\n")

    except pyodbc.Error as e:
        messagebox.showerror("Database Error", str(e))



def insert_data():
    try:
        table_name = entry_table_name_insert.get()
        values = entry_values.get()

        if not table_name or not values:
            messagebox.showwarning("Input Error", "Both table name and values are required!")
            return

       
        query = f"INSERT INTO {table_name} VALUES ({values})"
        cursor.execute(query)
        connection.commit()

        messagebox.showinfo("Success", f"Data inserted into table '{table_name}' successfully!")

    except pyodbc.Error as e:
        messagebox.showerror("Database Error", str(e))


def get_customer_info():
    try:
        customer_id = entry_customer_id.get()

        if not customer_id:
            messagebox.showwarning("Input Error", "Customer ID is required!")
            return

      
        cursor.execute("EXEC get_costumer_info @CostumerID = ?", customer_id)
        rows = cursor.fetchall()

        if not rows:
            messagebox.showinfo("No Data", f"No data found for Customer ID {customer_id}.")
            return

     
        data_window = tk.Toplevel()
        data_window.title(f"Customer Info for ID {customer_id}")

        text = tk.Text(data_window, wrap="word", width=80, height=20)
        text.pack()

  
        column_headers = ["CostumerID", "FirstName", "LastName", "Email", "OrderID", "ProductName"]
        text.insert("1.0", "\t".join(column_headers) + "\n")
        text.insert("2.0", "-" * 80 + "\n")


        for row in rows:
            text.insert("end", "\t".join(map(str, row)) + "\n")

    except pyodbc.Error as e:
        messagebox.showerror("Database Error", str(e))



root = tk.Tk()
root.title("Database Management System")


tk.Label(root, text="Retrieve Data", font=("Arial", 14, "bold")).grid(row=0, column=0, columnspan=2, pady=10)
tk.Label(root, text="Table Name:").grid(row=1, column=0, sticky="e")
entry_table_name = tk.Entry(root)
entry_table_name.grid(row=1, column=1)

tk.Button(root, text="Retrieve Data", command=retrieve_data).grid(row=2, column=0, columnspan=2, pady=10)


tk.Label(root, text="Insert Data", font=("Arial", 14, "bold")).grid(row=3, column=0, columnspan=2, pady=10)
tk.Label(root, text="Table Name:").grid(row=4, column=0, sticky="e")
entry_table_name_insert = tk.Entry(root)
entry_table_name_insert.grid(row=4, column=1)

tk.Label(root, text="Values (comma-separated):").grid(row=5, column=0, sticky="e")
entry_values = tk.Entry(root, width=50)
entry_values.grid(row=5, column=1)

tk.Button(root, text="Insert Data", command=insert_data).grid(row=6, column=0, columnspan=2, pady=10)


tk.Label(root, text="Get Customer Info", font=("Arial", 14, "bold")).grid(row=7, column=0, columnspan=2, pady=10)
tk.Label(root, text="Customer ID:").grid(row=8, column=0, sticky="e")
entry_customer_id = tk.Entry(root)
entry_customer_id.grid(row=8, column=1)

tk.Button(root, text="Get Info", command=get_customer_info).grid(row=9, column=0, columnspan=2, pady=10)

root.mainloop()

connection.close()
```
